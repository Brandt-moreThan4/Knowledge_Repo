```python
from typing import Iterable

def make_frame(vals:Iterable, row_length:int=7) -> pd.DataFrame:
    col_length = len(vals)//row_length+1
    array = np.zeros(shape=(row_length,col_length))
    array[:,:] = np.nan
    array = array.tolist()
    
    val_counter = 0
    for col in range(col_length):
        for row in range(row_length):
            array[row][col] = vals[val_counter]
            val_counter += 1
            if val_counter == len(vals):
                return pd.DataFrame(array)
```

```python
msft = yf.Ticker('msft')
```
